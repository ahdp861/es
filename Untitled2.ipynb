```python
import pandas as pd
import numpy as np
from collections import Counter

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def train_logistic_regression(X, y, learning_rate=0.01, n_iterations=1000):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0
    
    for _ in range(n_iterations):
        linear_model = np.dot(X, weights) + bias
        y_predicted = sigmoid(linear_model)
        
        dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
        db = (1 / n_samples) * np.sum(y_predicted - y)
        
        weights -= learning_rate * dw
        bias -= learning_rate * db
    
    return weights, bias

def predict_logistic_regression(X, weights, bias, threshold=0.5):
    linear_model = np.dot(X, weights) + bias
    probabilities = sigmoid(linear_model)
    return (probabilities >= threshold).astype(int)

def entropy(y):
    counts = np.bincount(y)
    probabilities = counts / len(y)
    return -np.sum([p * np.log2(p) for p in probabilities if p > 0])

def information_gain(X_column, y, threshold):
    left_mask = X_column <= threshold
    right_mask = X_column > threshold
    
    if np.sum(left_mask) == 0 or np.sum(right_mask) == 0:
        return 0
    
    parent_entropy = entropy(y)
    n = len(y)
    n_left, n_right = np.sum(left_mask), np.sum(right_mask)
    
    child_entropy = (n_left / n) * entropy(y[left_mask]) + (n_right / n) * entropy(y[right_mask])
    
    return parent_entropy - child_entropy

def best_split(X, y):
    best_gain = -1
    best_feature = None
    best_threshold = None
    
    for feature_idx in range(X.shape[1]):
        unique_values = np.unique(X[:, feature_idx])
        for threshold in unique_values:
            gain = information_gain(X[:, feature_idx], y, threshold)
            if gain > best_gain:
                best_gain = gain
                best_feature = feature_idx
                best_threshold = threshold
    
    return best_feature, best_threshold

def build_tree(X, y, max_depth=5, depth=0):
    if depth >= max_depth or len(np.unique(y)) == 1:
        return {'leaf': True, 'prediction': Counter(y).most_common(1)[0][0]}
    
    feature, threshold = best_split(X, y)
    
    if feature is None:
        return {'leaf': True, 'prediction': Counter(y).most_common(1)[0][0]}
    
    left_mask = X[:, feature] <= threshold
    right_mask = X[:, feature] > threshold
    
    left_tree = build_tree(X[left_mask], y[left_mask], max_depth, depth + 1)
    right_tree = build_tree(X[right_mask], y[right_mask], max_depth, depth + 1)
    
    return {
        'leaf': False,
        'feature': feature,
        'threshold': threshold,
        'left': left_tree,
        'right': right_tree
    }

def predict_tree_single(tree, x):
    if tree['leaf']:
        return tree['prediction']
    
    if x[tree['feature']] <= tree['threshold']:
        return predict_tree_single(tree['left'], x)
    else:
        return predict_tree_single(tree['right'], x)

def predict_decision_tree(tree, X):
    return np.array([predict_tree_single(tree, x) for x in X])

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def predict_knn_single(X_train, y_train, x_test, k=3):
    distances = [euclidean_distance(x_train, x_test) for x_train in X_train]
    k_indices = np.argsort(distances)[:k]
    k_labels = [y_train[i] for i in k_indices]
    return Counter(k_labels).most_common(1)[0][0]

def predict_knn(X_train, y_train, X_test, k=3):
    return np.array([predict_knn_single(X_train, y_train, x_test, k) for x_test in X_test])



data = pd.read_csv('telecom_churn (1).csv')

numerical_cols = data.select_dtypes(include=[np.number]).columns
numerical_cols = numerical_cols.drop('churn') if 'churn' in numerical_cols else numerical_cols

X = data[numerical_cols].values
y = data['churn'].values

def standardize(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    return (X - mean) / (std + 1e-8)

X = standardize(X)

def train_test_split_manual(X, y, test_size=0.25, random_state=42):
    np.random.seed(random_state)
    n_samples = len(X)
    indices = np.random.permutation(n_samples)
    split_idx = int(n_samples * (1 - test_size))
    train_idx, test_idx = indices[:split_idx], indices[split_idx:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

X_train, X_test, y_train, y_test = train_test_split_manual(X, y, test_size=0.25)

print("1. Логистическая регрессия:")
weights, bias = train_logistic_regression(X_train, y_train)
y_pred_lr = predict_logistic_regression(X_test, weights, bias)
acc_lr = np.mean(y_pred_lr == y_test)
print(f"   Accuracy: {acc_lr:.4f}")

print("\n2. Решающее дерево:")
tree = build_tree(X_train, y_train, max_depth=5)
y_pred_dt = predict_decision_tree(tree, X_test)
acc_dt = np.mean(y_pred_dt == y_test)
print(f"   Accuracy: {acc_dt:.4f}")

print("\n3. K ближайших соседей (k=3):")
y_pred_knn = predict_knn(X_train, y_train, X_test, k=3)
acc_knn = np.mean(y_pred_knn == y_test)
print(f"   Accuracy: {acc_knn:.4f}")

def confusion_matrix_manual(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    return np.array([[tn, fp], [fn, tp]])

accuracies = {'LogisticRegression': acc_lr, 'DecisionTree': acc_dt, 'KNN': acc_knn}
best_model = max(accuracies, key=accuracies.get)
print(f"\nЛучшая модель: {best_model} (accuracy = {accuracies[best_model]:.4f})")

if best_model == 'LogisticRegression':
    best_pred = y_pred_lr
elif best_model == 'DecisionTree':
    best_pred = y_pred_dt
else:
    best_pred = y_pred_knn

cm = confusion_matrix_manual(y_test, best_pred)
print(f"\nМатрица ошибок для {best_model}:")
print(cm)
```
